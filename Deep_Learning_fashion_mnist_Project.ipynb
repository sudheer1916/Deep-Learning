```python
#Normalization
X_train = X_train/255.0
X_test = X_test/255.0
#print(X_train[0])
```

```python
from keras import Sequential
from keras.layers import Dense, Flatten, Dropout
from keras.initializers import HeNormal
from keras.optimizers import Adam
```

```python
model = Sequential()

model.add(Flatten(input_shape = (28,28)))
print(model.output_shape)

model.add(Dense(128, activation='relu'))      #28*28*128 + 128
print(model.output_shape)

model.add(Dense(10, activation='softmax'))    #128*10 + 10
print(model.output_shape)
```

```python
model.summary()
```

```python
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer= optimizer,
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
model.fit(X_train,
          y_train,
          epochs=20,
          batch_size=64,
          verbose=True
          )
#default batch size = 32
# epoch = Forward pass + Backward pass
# Forward pass assign weights randomly at intial state
# Backward pass update the weight by ADAM
# loss is not reducing
# weights are same ====== vansihing gradients problem
# loss is reducing
# weights are same ==== we are good
# 60000/64= 938 batches
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test)
print(test_acc)
print(test_loss)

# Underfit: Train error high
# Normal fit: Train error low- test error low
# Overfit : Train errir low - Test error hig
```

```python
%%writefile app.py
import streamlit as st
import numpy as np
from PIL import Image
from keras import Sequential
from keras.layers import Dense, Flatten, Dropout
from keras.initializers import HeNormal
from keras.optimizers import Adam

st.title('upload image..!')
upload = st.file_uploader("upload a Image", type=['jpg', 'jpeg', 'png'])

def load_model():
  model = Sequential()
  model.add(Flatten(input_shape = (28,28)))
  model.add(Dense(128, activation='relu'))      #28*28*128 + 128
  model.add(Dense(10, activation='softmax'))    #128*10 + 10
  model.load_weights('./contents/weights')
  return model

model = load_model()

if upload is not None:
  img = Image.open(upload)
  st.image(img)
  click = st.button("Detect")
  if click:
    class_names = ['top', 'trouser', 'pullover', 'dress', 'coat', 'sandal', 'shirt',
              'sneaker', 'bag', 'ankle boot']
    resized_image = img.resize((28, 28)).convert('L')
    arr = np.array(resized_image)
    arr = arr/255.0
    pred = model.predict(arr.reshape(-1, 28 , 28))
    st.write(f"Prediction: {class_names[np.argmax(pred)]}")
```
